# YOLOv8 Trainingsprozess mit unterschiedlichen Parameter und Models(Objekterkennung)

## YOLOv8 Vergleich mit unterschiedlichen Models

## Model: yolov8n, yolov8s, yolov8m
### Parameter: 
* epochs = 200
* batch size = 16
* patience = 0
* iou = 0.1

### YOLOv8 mit Model yolos8n

```python
from ultralytics import YOLO
if __name__ == '__main__':
    model = YOLO("yolov8n.pt")
    model.train(data='/home/hiwis/Documents/gefertec_yolov8/yolov8_erkennung/erkennung/data.yaml', 
                epochs = 200, 
                batch = 16, 
                patience=0,
                device = 1
               )
```

### mAp Evaluation

```python
from ultralytics import YOLO
if __name__ == '__main__':
    model = YOLO('/home/hiwis/Documents/gefertec_yolov8/runs/detect/train2/weights/best.pt')  # load a custom model

# Validate the model
    metrics = model.val(iou=0.1) # no arguments needed, dataset and settings remembered
    print(f'mAP von Erkennung ist: {metrics}')
```
